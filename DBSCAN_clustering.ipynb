```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=250,noise=0.05)
```

```python
x
```

```python
y
```

```python
df={"data1":x[:,0],"data2":x[:,1]}
```

```python
dataset=pd.DataFrame(df)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset)
plt.show()
```

```python
from sklearn.cluster import DBSCAN 
```

```python
db=DBSCAN(eps=0.2,min_samples=5)
dataset["Predict"]=db.fit_predict(dataset)
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset,hue="Predict")
plt.show()
```

```python
# KMeans and hierarical clustering are used for linear data while DBSCAN clustering is used for non-linear data
```
