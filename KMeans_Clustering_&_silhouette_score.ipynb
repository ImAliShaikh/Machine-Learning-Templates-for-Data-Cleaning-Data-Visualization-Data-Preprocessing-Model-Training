```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
dataset=pd.read_csv(r"Iris.csv")
dataset.head(3)
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(2,21):
    km=KMeans(n_clusters=i,init="k-means++")
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
plt.plot([i for i in range(2,21)],wcss,marker="o")
plt.xlabel("no of cluste")
plt.xticks([i for i in range(2,21)])
plt.ylabel("WCSS")
plt.grid(axis="x")
plt.show()
```

```python
from sklearn.cluster import KMeans


kmn = KMeans(n_clusters=3)
dataset["Predict"] = kmn.fit_predict(dataset)
```

```python
from sklearn.metrics import silhouette_score
kmn.labels_
# silhouette score
```

```python
silhouette_score(dataset,labels=kmn.labels_)
```

```python
ss=[]
no_c=[j for j in range(2,21)]
for i in range(2,21):
    km1=KMeans(n_clusters=i)
    km1.fit(dataset)
    ss.append(silhouette_score(dataset,labels=km1.labels_))
```

```python
plt.plot(no_c,ss)
plt.xlabel("no of cluster")
plt.ylabel("silhouette score")
plt.xticks(no_c)
plt.grid(axis="x")
plt.show()
```
