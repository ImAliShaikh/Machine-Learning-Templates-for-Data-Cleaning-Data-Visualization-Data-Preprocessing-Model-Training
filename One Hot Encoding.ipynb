# One hot Encoding and Dummy Variables

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
dataset=pd.read_csv("C:\\Users\\HP\\Documents\\Machine Learning\\loan 3 (1).csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True)
```

```python
dataset["Married"].fillna(dataset["Married"].mode()[0],inplace=True)
```

```python
# encoding
en_data=dataset[["Gender","Married"]]
en_data
```

```python
pd.get_dummies(en_data)
```

```python
# Another Method
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
ar=ohe.fit_transform(en_data).toarray()
# this is in array form
# fir_transform observes data and convert it from categorical data to numerical data directly
```

```python
pd.DataFrame(ar,columns=["Gender_Female","Gender_Male","Married_No","Married_Yes"])
```

```python
# Another short method
ohe=OneHotEncoder(drop="first")
ar=ohe.fit_transform(en_data).toarray()
pd.DataFrame(ar,columns=["Gender_Male","Married_Yes"])
```
